1. What is a random variable in probability theory?

  - A random variable is a numerical outcome of a random process or experiment. It assigns values to outcomes in a sample space.

2. What are the types of random variables?

  - Discrete: Takes specific values (e.g, number of heads in 3 coin tosses).

  - Continuous: Takes any value in a range (e.g, height, weight, time).

3. What is the difference between discrete and continuous distributions?

  - Discrete distributions describe variables with countable outcomes.

  - Continuous distributions describe variables with infinite, uncountable outcomes.

4. What are probability distribution functions (PDF)?

  - For discrete variables, it's the Probability Mass Function (PMF).

  - For continuous variables, it's the Probability Density Function (PDF), showing likelihood over intervals.

5. How do cumulative distribution functions (CDF) differ from PDFs?

  - CDF gives the probability that a variable is less than or equal to a value. PDF gives the probability density at a specific value (for continuous) or the exact probability (for discrete).

6. What is a discrete uniform distribution?

  - All outcomes have equal probability. Example: rolling a fair die (1–6, each has P = 1/6).

7. What are the key properties of a Bernoulli distribution?

  - Only two outcomes: success (1) and failure (0).

  - One trial.

  - Probability of success = p, failure = 1 - p.

8. What is the binomial distribution and how is it used in probability?

  - Models number of successes in n independent Bernoulli trials. Useful for yes/no experiments repeated multiple times.

9. What is the Poisson distribution and where is it applied?

  - Models the number of events in a fixed interval of time/space with a known constant rate. Used in queuing theory, call centers, etc.

10. What is a continuous uniform distribution?

  - All values within a certain interval are equally likely. Example: random real number between 0 and 1.

11. What are the characteristics of a normal distribution?

  -Characteristics of a normal distribution are:  
  
  * Symmetrical bell shape.

  * Mean = median = mode.

  * Defined by mean (μ) and standard deviation (σ).

12. What is the standard normal distribution, and why is it important?
  
  - A normal distribution with μ = 0, σ = 1. Used to compute Z-scores and standardize data for comparison.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  
  - CLT states that the sampling distribution of the sample mean approximates a normal distribution as sample size increases, regardless of the original distribution. It's key for inference.

14. How does the Central Limit Theorem relate to the normal distribution?
  
  - It explains why normal distribution appears in many natural phenomena and justifies using normal-based methods for sample means.

15. What is the application of Z statistics in hypothesis testing?
  
  - Z-tests compare a sample statistic to a known population parameter under normality assumptions. Common in large-sample tests.

16. How do you calculate a Z-score, and what does it represent?
  
  - Z = (X - μ) / σ
  It shows how many standard deviations a value is from the mean.


17. What are point estimates and interval estimates in statistics?

  - Point estimate: Single value (e.g, sample mean) as best guess.

  - Interval estimate: Range of values (e.g, confidence interval) likely to contain the parameter.

18. What is the significance of confidence intervals in statistical analysis?

  - They provide a range where the true population parameter is expected to lie, with a specified level of confidence (e.g., 95%).

19. What is the relationship between a Z-score and a confidence interval?

  - Z-scores are used to calculate the margin of error in confidence intervals when the population standard deviation is known.

20. How are Z-scores used to compare different distributions?

  - Z-scores standardize values, making it possible to compare data from different distributions or units.

21. What are the assumptions for applying the Central Limit Theorem?

  - Random sampling.

  - Independent observations.

  - Sample size is large (n ≥ 30) or population is normally distributed.

22. What is the concept of expected value in a probability distribution?

  - Expected value is the long-run average outcome of a random variable, calculated as E(X) = Σ[p(x) * x].

23. How does a probability distribution relate to the expected outcome of a random variable?

  - The probability distribution shows all possible outcomes and their probabilities, and the expected value summarizes the average of those outcomes over many trials.
